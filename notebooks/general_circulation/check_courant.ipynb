```python
import numpy as np
import numpy.ma as ma
import netCDF4 as nc
%matplotlib inline 
import matplotlib.pyplot as plt
from salishsea_tools import diagnosis_tools
import glob

src = '/ocean/imachuca/Canyons/results_mackenzie/ideal_double_flt/'

up = glob.glob(src + 'MCKNZ*1h*U*.nc')
vp = glob.glob(src + 'MCKNZ*1h*V*.nc')
wp = glob.glob(src + 'MCKNZ*1h*W*.nc')
mesh_mask_p = nc.Dataset(src + 'mesh_mask.nc')

uc = glob.glob(src + '1_MCKNZ*1h*U*.nc')
vc = glob.glob(src + '1_MCKNZ*1h*V*.nc')
wc = glob.glob(src + '1_MCKNZ*1h*W*.nc')
mesh_mask_c = nc.Dataset(src + '1_mesh_mask.nc')
```

```python
cuc = diagnosis_tools.pcourantu(uc, mesh_mask_c)
print("cuc")
cvc = diagnosis_tools.pcourantv(vc, mesh_mask_c)
print("cvc")
cwc = diagnosis_tools.pcourantw(wc, mesh_mask_c)
print("cwc")

cup = diagnosis_tools.pcourantu(up, mesh_mask_p)
print("cup")
cvp = diagnosis_tools.pcourantv(vp, mesh_mask_p)
print("cvp")
cwp = diagnosis_tools.pcourantw(wp, mesh_mask_p)
print("cwp")
```

```python
plt.figure(figsize=(16,6))
dt=240
plt.subplot(1,3,1); plt.pcolormesh(cup*dt); plt.colorbar()
plt.subplot(1,3,2); plt.pcolormesh(cvp*dt); plt.colorbar()
plt.subplot(1,3,3); plt.pcolormesh(cwp*dt); plt.colorbar()
```
